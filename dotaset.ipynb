```python
# Из всего датасета с kaggle выбрали только нужные нам csv файлы, которые описывают по id матча победителя,
# также по id матча на каких героях играли игроки и id героев сопоставили имена героев.
# Также огромный минус данного датасета, что среди таблиц нет информации о рейтинге игроков, т.е. неизвестно, на какком рейтинге происходят игры
test_players = pd.read_csv('./test_player.csv')
hero_names = pd.read_csv('./hero_names.csv')
test_labels = pd.read_csv('./test_labels.csv', index_col=0)
hero_names_players = pd.merge(test_players, hero_names, on ='hero_id', how ='left')
hero_names_players.set_index(['match_id', 'player_slot'], inplace = True)
heroes = hero_names_players['hero_id'].unstack()
name_heroes_players = hero_names_players['localized_name'].unstack()
data = pd.merge(test_labels, heroes, left_index = True, right_index = True)
data = pd.merge(data, name_heroes_players, left_index = True, right_index = True)
data = data.drop(['0_x', '1_x', '2_x', '3_x', '4_x', '128_x', '129_x', '130_x', '131_x', '132_x'], axis=1)
data = data.rename(columns={'0_y': '0', '1_y': '1', '2_y': '2', '3_y': '3', '4_y': '4', '128_y': '5', '129_y': '6', '130_y': '7', '131_y': '8', '132_y': '9'})
# Тут скачиваем наш обработанный с kaggle датасет
data.to_csv('dotaset_kaggle.csv', index=False)
data.head()
```
